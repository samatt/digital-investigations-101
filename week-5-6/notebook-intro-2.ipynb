```python
import pandas as pd
import altair as alt
import datetime
```

```python
restaurants = pd.read_csv("DOHMH_New_York_City_Restaurant_Inspection_Results.csv")
```

```python
restaurants.head()
```

```python
boros = restaurants.groupby("BORO", dropna = False).size().to_frame('size').reset_index()
```

```python
boros
```

```python
boros = boros[boros.BORO != "0"]
```

```python
alt.Chart(boros).mark_bar().encode(
    y="BORO",
    x="size"
).properties(
    title="Boro rest count"
)
```

```python
boros_cuisine = restaurants.groupby(["CUISINE DESCRIPTION", "BORO"], dropna = False).size().to_frame('size').reset_index().sort_values("size", ascending = False)
boros_cuisine = boros_cuisine[(boros_cuisine.BORO != "0") & (boros_cuisine.size > 100)]
```

```python
alt.Chart(boros_cuisine[boros_cuisine.BORO == "Manhattan"]).mark_bar().encode(
    x= alt.X("CUISINE DESCRIPTION", sort = "-y"),
    y= "sum(size)",
    # facet = "BORO"
).properties(
    title="Manhattan cuisine count"
)
```

```python
alt.Chart(boros_cuisine[boros_cuisine.BORO == "Brooklyn"]).mark_bar().encode(
    x= alt.X("CUISINE DESCRIPTION", sort = "-y"),
    y= "sum(size)",
    # facet = "BORO"
).properties(
    title="Brooklyn cuisine count"
)
```

```python
alt.Chart(boros_cuisine[boros_cuisine.BORO == "Bronx"]).mark_bar().encode(
    x= alt.X("CUISINE DESCRIPTION", sort = "-y"),
    y= "sum(size)",
    # facet = "BORO"
).properties(
    title="Bronx cuisine count"
)
```

```python
rests_by_date = restaurants[["BORO", "DBA", "ZIPCODE", "INSPECTION DATE"]]
rests_by_date = rests_by_date.groupby(["BORO", "INSPECTION DATE"]).size().to_frame('size').reset_index().sort_values("size", ascending = False)
rests_by_date['INSPECTION DATE'] = pd.to_datetime(rests_by_date['INSPECTION DATE'])
rests_by_date =rests_by_date[(rests_by_date['INSPECTION DATE'] > '2018-01-01') ]
```

```python
alt.Chart(rests_by_date).mark_bar().encode(
    x ="yearmonth(INSPECTION DATE)",
    y="size",
    # facet="BORO",
)
```

```python
alt.Chart(rests_by_date).mark_bar().encode(
    x ="yearmonth(INSPECTION DATE)",
    y="size",
    facet="BORO",
)
```

```python
#CUISINE DESCRIPTION
rests_by_date = restaurants[["BORO", "DBA", "ZIPCODE", "INSPECTION DATE", "GRADE"]]
rests_by_date = rests_by_date.groupby(["BORO", "INSPECTION DATE", "GRADE"]).size().to_frame('size').reset_index().sort_values("size", ascending = False)
rests_by_date['INSPECTION DATE'] = pd.to_datetime(rests_by_date['INSPECTION DATE'])
rests_by_date =rests_by_date[(rests_by_date['INSPECTION DATE'] > '2019-01-01') ]
```

```python
rests_by_date = rests_by_date[rests_by_date.BORO != "0"]
```

```python
alt.data_transformers.disable_max_rows()
```

```python
alt.Chart(rests_by_date).mark_bar().encode(
    x ="yearmonth(INSPECTION DATE)",
    y="size",
    facet="BORO",
    color="GRADE",
).interactive()
```
